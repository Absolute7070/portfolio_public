### Check equilibrium 

```python
potentials = np.loadtxt("../10nm/potential.xvg", skiprows=24)
plt.plot(potentials[:, 0],potentials[:, 1], '.' )
plt.xlabel("t (ps)")
plt.ylabel("potential energy (kJ/mol)")


plt.axvline(x = 300, color = 'r', label = 'axvline - full height')
```

After about 300 ps, we achieve equilibrium. 

### Computing surface tension

Since after 300ps, we achieved equilibrium, we only extract those values of instantaneous surface tension and average it to get our surface tension result. 

```python
pressures =  np.loadtxt("../10nm/pressure_diagonals.xvg", skiprows=26) # time (ps), pressure (bar) 
pressures
```

Now the surface is on xz-plane. Then surface tension is
$$
\gamma = \frac{L_z}{2} (P_{yy}- \frac{P_{zz}+ P_{xx}}{2})
$$

```python
L_z =  10 # nm 
```

```python
surface_tension = L_z/2 * (pressures[:, 2]-  (pressures[:, 1]+ pressures[:, 3])/2 )
```

```python
# plot surface tension 
plt.plot(pressures[:, 0], surface_tension, '.'); 
```

```python
surface_tension_equil= surface_tension[300:] # instantaneous surface tension values after equlibrium achived
```

```python
surface_tension_equil.mean()
```

```python
print("Last instantaneous surface tension:", surface_tension_equil[-1])   # 
print("Mean surface tension during equilibrium:", surface_tension_equil.mean())
```
